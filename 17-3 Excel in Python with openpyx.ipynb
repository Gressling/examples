```
from openpyxl import Workbook
wb = Workbook()
```

```
# active worksheet
ws = wb.active
# Data can be assigned directly to cells
ws['A1'] = 'Ethanol'
ws['B1'] = 'CCO'
# Rows can also be appended
ws.append([1, 2, 3])
# Calculations and formula
ws["A3"] = "=LEN(B1)"
```

```
# Save the file
wb.save("smiles.xlsx")
```
