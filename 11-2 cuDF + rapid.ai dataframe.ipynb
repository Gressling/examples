```
import pandas as molCoord
import numpy as np
import cudf

pandas_df = molCoord.DataFrame(
 {'x': np.random.randint(0, 100000, size=100000),
 'y': np.random.randint(0, 100000, size=100000),
 'z': np.random.randint(0, 100000, size=100000)})

cudf_df = cudf.DataFrame.from_pandas(pandas_df)

# Timing Pandas
%timeit pandas_df.a.mean()
# Timing cuDF
%timeit cudf_df.a.mean() # 16x times faster with 1080 Ti GPU, CUDA 10
```
