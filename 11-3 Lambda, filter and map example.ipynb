```
x = ['Python', 'programming', 'for', 'data science!']
print(sorted(x))
```

```
print(sorted(x, key=lambda arg: arg.lower()))
```

```
# filter
print(list(filter(lambda arg: len(arg) < 8, x)))
```

```
print(list(map(lambda arg: arg.upper(), x)))
```
